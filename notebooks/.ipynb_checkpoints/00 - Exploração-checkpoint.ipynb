# 00 - Bibliotecas

```python
import kaggle
from kaggle.api.kaggle_api_extended import KaggleApi
import zipfile
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 01 - Importa√ß√£o de Dados

```python
# Configura√ß√£o de caminhos
caminho_pasta_dados = '../Dados/'

# Criar pasta Dados se n√£o existir
os.makedirs(caminho_pasta_dados, exist_ok=True)
print(f"‚úÖ Pasta de dados configurada: {caminho_pasta_dados}")
```

```python
# Download do dataset do Kaggle
print("‚¨áÔ∏è Iniciando download do dataset...")

try:
    api = KaggleApi()
    api.authenticate()
    api.dataset_download_files(
        'saurabhbadole/bank-customer-churn-prediction-dataset', 
        path=caminho_pasta_dados
    )
    print("‚úÖ Download conclu√≠do com sucesso!")
except Exception as e:
    print(f"‚ùå Erro no download: {e}")
```

```python
# Extra√ß√£o do arquivo ZIP
caminho_zip = os.path.join(caminho_pasta_dados, 'bank-customer-churn-prediction-dataset.zip')

print("‚¨áÔ∏è Extraindo arquivos...")
try:
    with zipfile.ZipFile(caminho_zip, 'r') as zip_ref:
        zip_ref.extractall(caminho_pasta_dados)
    print(f"‚úÖ Dataset extra√≠do em: {caminho_pasta_dados}")
except Exception as e:
    print(f"‚ùå Erro na extra√ß√£o: {e}")
```

```python
# Verifica√ß√£o dos arquivos baixados
print("\nüìÅ Conte√∫do da pasta Dados:")
arquivos = os.listdir(caminho_pasta_dados)
for i, arquivo in enumerate(arquivos, 1):
    print(f"  {i}. üìÑ {arquivo}")
```

```python
# Carregar o dataset
try:
    # Tenta encontrar o arquivo CSV (pode ter nome diferente)
    arquivos_csv = [f for f in arquivos if f.endswith('.csv')]
    
    if arquivos_csv:
        caminho_dataset = os.path.join(caminho_pasta_dados, arquivos_csv[0])
        df = pd.read_csv(caminho_dataset)
        print(f"‚úÖ Dataset carregado: {arquivos_csv[0]}")
        print(f"üìä Dimens√µes: {df.shape[0]} linhas √ó {df.shape[1]} colunas")
    else:
        print("‚ùå Nenhum arquivo CSV encontrado!")
        
except Exception as e:
    print(f"‚ùå Erro ao carregar dataset: {e}")
```

```python
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1, inplace = True)
df.head()
```

```python
colunas_categoricas = df.select_dtypes(include=['object', 'category']).columns
colunas_numericas = df.select_dtypes(include=[np.number]).columns

print(f"Colunas categ√≥ricas: {colunas_categoricas}")
print(f"Colunas num√©ricas: {colunas_numericas}")
```

```python
df.describe()
```

```python
df[colunas_categoricas].value_counts()
```

```python
colunas_para_plot = colunas_numericas.drop(['HasCrCard', 'IsActiveMember', 'Exited'])
n_colunas_para_plot = len(colunas_para_plot)

ncols = 3
nrows = (n_colunas_para_plot + 2) // ncols

fig, axes = plt.subplots(nrows = nrows, ncols = ncols, figsize=(15, 5 * nrows))
axes = axes.flatten()
for i, coluna in enumerate(colunas_para_plot):
    sns.boxplot(data = df, y = coluna, hue = 'Exited', gap=.2, width = 0.5, ax = axes[i])
    axes[i].set_title(f'Boxplot de {coluna} por Exited')
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)
plt.tight_layout()
plt.show()
```
